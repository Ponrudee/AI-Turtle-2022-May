##Plot Image

```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2

#extracted frame frome 24fps video
vidcap = cv2.VideoCapture('24.00fps-TargetFPS-mode3-rife-Video.mp4')
success,image = vidcap.read()
count = 0
path = './ExtractedFrames'
if not os.path.exists(path):
  os.makedirs(path)
while success:
  cv2.imwrite(os.path.join(path , '%d.jpg')%count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  #print('Read a new frame: ', success)
  count += 1


n_original_frame = 4
#plot 24fps figure
plt.figure(figsize=(8*2, 4*2))
for i in range((n_original_frame-1)*8 + 1):
    plt.subplot(4, 8, i+1)
    plt.gca().set_axis_off()
    plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
    plt.margins(0,0)
    img = Image.open('./ExtractedFrames/%d.jpg'%i)
    plt.imshow(img)
    plt.axis('off')
plt.show()

#plot original frame figure
plt.figure(figsize=(1*2, 4*2))
for i in range(1, n_original_frame+1):
    plt.subplot(4, 1, i)
    plt.gca().set_axis_off()
    plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
    plt.margins(0,0)
    img = Image.open('../Result/%d.jpg'%i)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```
