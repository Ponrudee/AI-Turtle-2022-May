```python
from PIL import Image

rootdir = 'Animation/NewEdition/DATATURTLECOLOR/png/'
output_path = 'Animation/NewEdition/DATATURTLECOLOR/jpg/'
n = 1

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
      png = Image.open(os.path.join(subdir, file))
      png.load() # required for png.split()

      jpg = Image.new("RGB", png.size, (255, 255, 255))
      jpg.paste(png, mask=png.split()[3]) # 3 is the alpha channel

      savename = '%d.jpg'%n
      if not os.path.exists(output_path):
            os.makedirs(output_path)
      jpg.save(output_path + savename, 'JPEG', quality=100, subsampling=0)
      n = n+1
```
